```python
##Opening CSV file through URL
import requests
from io import StringIO
url = "https://raw.githubusercontent.com/plotly/datasets/refs/heads/master/data.csv"
headers = {"User-Agent": "Mozilla/5.0(Macintosh; Intel Mac OS X 10.14; rv: 66.0)Gecko/20100101 Firefox/66.0"}
req = requests.get(url,headers = headers)
data = StringIO(req.text)
df2 = pd.read_csv(data)
df2.shape
```

```python
##Index_col parameter, converts col into index
pd.read_csv('placement.csv', index_col = 'Unnamed: 0')
```

```python
#Header Parameter, when u have cols name in the first row then
pd.read_csv('placement.csv', header = 1)
```

```python
##Use_cols parameter
pd.read_csv('placement.csv', usecols = ['cgpa', 'iq', 'placement'])
```

```python
##Skiprows
pd.read_csv('placement.csv', skiprows= [0, 3])
```

```python
##N rows
pd.read_csv('placement.csv', nrows = 10)
```

```python
#Encoding parameters of any thing other then UTF8
pd.read_csv('zomato.csv', encoding = "latin-1")
```

```python
#Skip bad lines
pd.read_csv('zomato.csv', encoding = "latin-1", on_bad_lines = 'skip')
```

```python
#dtypes parameter
pd.read_csv('placement.csv', dtype = {'iq': int})
```

```python
#Handling Dates
pd.read_csv('matches.csv', parse_dates = ['date']).info()
##if u use parse_dates= [[1,3]] can combine 1, 3
```

```python
##Convertos
def rename(name):
    if name == 'Mumbai Indians':
        return 'MI'
    else :
        return name
pd.read_csv('matches.csv', converters = {'team1' : rename})
```

```python
##na_values parameter, consider any values as NaN
pd.read_csv('matches.csv', na_values = ['Delhi'])
```

```python
#Loading huge chunks of data
dfs = pd.read_csv('titanic.tsv',sep = '\t', chunksize = 100)
for chunk in dfs:
    print(chunk.shape)
```
