```python
mask = ipl['batsman_runs'] == 4
fours = ipl[mask].groupby('batsman')
fours['batsman_runs'].count().sort_values(ascending = False).head()
```

```python
m = ipl['batsman'] == 'V Kohli'
mostRuns = ipl[m].groupby('bowling_team')
mostRuns['batsman_runs'].sum().sort_values(ascending = False)
```

```python
def mostRun(bman):
    m = ipl['batsman'] == bman
    mostRuns = ipl[m].groupby('bowling_team')
    return mostRuns['batsman_runs'].sum().sort_values(ascending = False).head(1)
mostRun('DA Warner')
```

```python
#isin function
```

```python
mostBalls = ipl.groupby('batsman')
mostBalls['batsman_runs'].count().sort_values(ascending = False).head(5)
```

```python
m = ipl['over'] > 15
death_overs = ipl[m].groupby('batsman')
ano = death_overs['batsman_runs'].count() 
x = ano > 200
batsman_list = ano[x].sort_values(ascending = False).index.tolist()
#Runs scored by this 43 player
m2 = ipl['batsman'].isin(batsman_list)
ano2 = ipl[m2].groupby('batsman')
runs = ano2['batsman_runs'].sum()
##Balls played by this 43 player
balls = ano2['batsman_runs'].count()
##STrike Rate
SR = (runs / balls) * 100
SR.sort_values(ascending = False).head(1)
```

```python
##Merge function
new = ipl.merge(data, left_on = 'match_id', right_on = 'id')
orange_cap = new.groupby(['season', 'batsman'])
orange_cap['batsman_runs'].sum().reset_index().sort_values(by = 'batsman_runs', ascending = False).drop_duplicates(subset = 'season', keep = 'first').sort_values(by = 'season')[['season', 'batsman']]
```

```python
#Pivot Table
mask2 = ipl['batsman_runs'] == 6
sixes = ipl[mask2]
pt = sixes.pivot_table(index = 'over', columns = 'batting_team', values = 'batsman_runs', aggfunc = 'count')
```
